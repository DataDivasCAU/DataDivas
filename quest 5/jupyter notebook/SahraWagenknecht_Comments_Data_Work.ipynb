# sorting comment Data

```python
import pandas as pd
```

## importing csv files

```python
df_After = pd.read_csv('SahraWagenknechtCommentsAfter.csv')
df_After
```

```python

df_Before = pd.read_csv('SahraWagenknechtCommentsBefore.csv')
df_Before
```

## sorting comments by likes, cutting of at the top 500 and saving the Dataframe as a csv

```python
df_Before_Popular = df_Before.sort_values(
    by='likeCount', ascending= False).head(500).reset_index(drop= True)
df_Before_Popular.to_csv('Popular_Comments_Before.csv')
df_Before_Popular
```

```python
df_After_Popular = df_After.sort_values(
    by='likeCount', ascending= False).head(500).reset_index(drop= True)
df_After_Popular.to_csv('Popular_Comments_After.csv')
df_After_Popular
```
